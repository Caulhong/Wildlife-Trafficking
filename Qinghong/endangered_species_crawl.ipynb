```python
from bs4 import BeautifulSoup
from selenium import webdriver

dr = webdriver.Chrome()
dr.get('https://cites.org/eng/app/appendices.php')
bs = BeautifulSoup(dr.page_source,"lxml")
```

```python
print(bs)
```

```python
def remove_content_between_parentheses(text):
    text = text.replace('\xa0', '')
    while '(' in text and ')' in text:
        start_index = text.find('(')
        end_index = text.find(')')
        
        if start_index != -1 and end_index != -1:
            text = text[:start_index] + text[end_index + 1:]

    return text
```

```python
x = bs.find("table", attrs={"class": "table table-hover table-striped"})
y = x.tbody.find_all("tr")
ans = []
for i in range(6, len(y)):
    tmp = y[i].find('div', class_='description')
    if tmp:
        temp = remove_content_between_parentheses(tmp.text).split(',')
        for j in temp:
            j = j.strip()
            if j and j != "etc.":
                ans.append(j)
print(ans)
```

```python
import csv
with open("Species.csv", mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Species"])
    for item in ans:
        writer.writerow([item])
```

```python
x = bs.find("table", attrs={"class": "table table-hover table-striped"})
y = x.tbody.find_all("tr")
res1 = []
for i in y:
    z = i.find_all("td")
    if(len(z) >= 2 and z[1].text and z[1].text != 'I'):
        res1.append(z[1].span.text)
print(res1)
```

```python
res2 = []
for i in y:
    z = i.find_all("td")
    if(len(z) >= 3 and z[2].text):
            z2 = z[2].find("class" == "entry entry-bundle-species_families entry-kingdom-258")
            z3 = z[3].find("class" == "entry entry-bundle-species_families entry-kingdom-258")
            if(z2):
                res2.append(z2.text)
            if(z3):
                res2.append(z3.text)
print(res2)
```

```python
import csv
with open("Species1.csv", mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Species"])
    for item in res1:
        writer.writerow([item])
with open("Species2.csv", mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Species"])
    for item in res2:
        writer.writerow([item])
```
